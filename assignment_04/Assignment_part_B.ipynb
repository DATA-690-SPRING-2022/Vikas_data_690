```python
# cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function

def cleanse_year(year_str):
  if "*" in year_str:
    year = year_str.replace("*", "")
  else:
    year = year_str

  return int(year)
```

```python
df["year_cleansed"] = df["Census Year"].apply(cleanse_year)
```

```python
def cleanse_pop(pop_str):
  if "," in pop_str:
    pop = pop_str.replace(",", "")
  else:
    pop = pop_str

  return int(pop)
```

```python
df["pop_cleansed"] = df["Total Population"].apply(cleanse_pop)
```

```python
def cleanse_cost(cost_str):
  data = cost_str.replace(",","").replace("$","").replace(" ","")
  if "Billion" in cost_str:
    data = data.replace("Billion","")
    data = float(data)*10000000
  
  return int(data)
```

```python
df["cost_cleansed"] = df["Census Cost"].apply(cleanse_cost)
```

```python
def cleanse_avg(avg_str):
  data = avg_str.replace("$","")
  if "cents" in data:
    data = data.replace("cents","")
    data = float(data)/100

  return float(data)
```

```python
df["avg_person"] = df["Average Cost Per Person"].apply(cleanse_avg)
df
```

```python
# Find out the summary statistics of the numerical columns - use describe() function
df.describe()
```

```python
# Use Pandas built-in plotting function to plot the average cost per person over time.

import matplotlib.pyplot as plt

x=df['year_cleansed']
y=df['avg_person']
plt.plot(x,y)
plt.xlabel('year_cleansed')
plt.ylabel('average cost per person')
plt.show()
```
