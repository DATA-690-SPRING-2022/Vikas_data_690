```python
import pandas as pd
import plotly.express as px
import plotly
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/saivikaskondamidi/Vikas_data_690/main/world_development_explorer/wdi_data-3.csv')
df.head()
```

```python
df.tail()
```

```python
df['Country Code'].unique()
```

```python
df['Country Name'].unique()
```

```python
df_Pop = df.query("indicator == 'SP.POP.TOTL' ")
df_Pop
```

```python
df_Pop = df.query("Year == 2016").query("indicator == 'SP.POP.TOTL'")

df_Pop_sorted = df_Pop.sort_values(by= "value", ascending= False)
df_Pop_sorted
```

#Finding which country has the highest and lowest Population by 2020

```python
fig = px.bar(
    data_frame= df_Pop_sorted,
    x= "Country Name",
    labels={"value":"2016 population, total"},
    y= "value",
    color= "Country Name",
    height= 700,
    template=list(plotly.io.templates.keys())[5],
    title= "TOP 10 countries by 2020 population, total"
)

fig.update_layout(showlegend= True)
fig.show()
```

```python
df_directors = df.query("indicator in(['IC.WEF.LLCD.MA'])")

df_directors
```

```python
df_directors = df.query("Year == 2016").query("indicator == 'IC.WEF.LLCD.MA' ")

df_directors_sorted = df_directors.sort_values(by= "value", ascending= False)
df_directors_sorted
```

#Finding Male Directors in the Private Sectors using data of 2 regions

```python
fig = px.pie(df_directors_sorted, values='value', names='Country Name', title='2016 number of male directors',template=list(plotly.io.templates.keys())[5])
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(showlegend=True)

fig.show()
```
